```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load one audio file (replace with actual path inside UrbanSound8K/fold1/)
y, sr = librosa.load("/content/UrbanSound8K/audio/fold1/7061-6-0-0.wav", duration=4)

# Listen (Colab can play audio!)
from IPython.display import Audio
Audio(y, rate=sr)
```

```python
# Extract MFCCs
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)

print("MFCC Shape:", mfcc.shape)

# Visualize MFCC
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar()
plt.title("MFCC Features")
plt.show()
```
